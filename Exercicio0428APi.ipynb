```python
import requests
import pandas as pd
import time

API_KEY = 'CK6BZ04TS4VFH9JC'

symbols = {
    'netflix': 'NFLX',
    'apple': 'AAPL',
    'google': 'GOOGL',
    'amazon': 'AMZN'
}

bases = {}

for nome, symbol in symbols.items():
    url = f'https://www.alphavantage.co/query?function=GLOBAL_QUOTE&symbol={symbol}&apikey={API_KEY}'
    r = requests.get(url)
    data = r.json()

    bases[nome] = data

    time.sleep(12)
```

```python
def criar_df(seguro_base, nome):
    if 'Global Quote' in seguro_base and seguro_base['Global Quote']:
        df = pd.DataFrame.from_dict([seguro_base['Global Quote']])
        df['empresa'] = nome
        return df
    else:
        print(f"Erro ao obter dados de {nome}: {seguro_base}")
        return pd.DataFrame()

base_alpha_netflix = criar_df(bases['netflix'], 'Netflix')
base_alpha_apple = criar_df(bases['apple'], 'Apple')
base_alpha_google = criar_df(bases['google'], 'Google')
base_alpha_amazon = criar_df(bases['amazon'], 'Amazon')

df_final = pd.concat([
    base_alpha_netflix,
    base_alpha_apple,
    base_alpha_google,
    base_alpha_amazon
], ignore_index=True)

df_final
```
